**What is Index**

-   So, indexing is nothing but allocating the name to each and every character which is in between the quotes.

-  we have 2 types of index

-  Forward Index -> Forward index will start from zero.
-  Backward Index -> Backward index starts from minus one.

```python
a='data'
print(a[0])
print(a[1])
print(a[2])
print(a[3])

print(a[-1])
print(a[-2])
print(a[-3])
```

```python
f = 'data science' # save in the Memory

print(f) # calling from the Memory

print(f[5] + f[6] + f[7])
```

**what is Slicing Operations**

-  Slicing operations in Python are used to access a range of characters from the memory using the variable.



Slicing Syntax : varibale[sp : ep]

```python
print(f[0] + f[5]  + f[10])

print(f[2] + f[3] + f[4] + f[5] + f[6])
```

```python
print(f[-11 : -3])
```

Skipping the Characters

Syntax : varibale[where to Start : where to end : step size ]

```python
print(f[:  :  2])
```

```python
print(f[:  :  3])
```

```python
print(f[2 : 10 : 2])
```

```python
print(f[:  :  -1])
```

**What is Mutable and Immutable**

-**  Mutable **: once the variable is saved in the temporary memory as a developer, if you are capable of increasing the memory size and decreasing the memory size, it is known as mutable.

-  **IMMutable**: once the variable is saved in the temporary memory as a developer, if you are not capable of increasing the memory size and decreasing the memory size, it is known as immutable.

```python
password = "Python123"

if (len(password) >= 8 and
    any(ch.isupper() for ch in password) and
    any(ch.islower() for ch in password) and
    any(ch.isdigit() for ch in password)):
    print("Valid Password")
else:
    print("Invalid Password")
```

```python
text = input("Enter a string: ")

vowels = 0
consonants = 0
digits = 0

for ch in text:
    if ch.lower() in "aeiou":
        vowels += 1
    elif ch.isalpha():
        consonants += 1
    elif ch.isdigit():
        digits += 1

print("Total Characters:", len(text))
print("Total Vowels:", vowels)
print("Total Consonants:", consonants)
print("Total Digits:", digits)
print("Reverse String:", text[::-1])
print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
```
