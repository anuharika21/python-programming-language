```python
def perfect(n):
  s=0
  for i in range(1,n):
    if n % i ==0:
      s=s+i
  if n==s:
    print(f"{n}:is perfect num")
  else:
    print(f"{n}: is not perfect")

perfect(6)
```

```python
#Calling a function inside another function
def is_even(x):
  if x%2==0:
    return True
  else:
    False

# div function to check divisibility
def div(y):
  if is_even(y) and y%3==0:
    return True
  else:
    return False


print(div(6))
```
