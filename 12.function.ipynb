```python
def perfect(n):
  s=0
  for i in range(1,n):
    if n % i ==0:
      s=s+i
  if n==s:
    print(f"{n}:is perfect num")
  else:
    print(f"{n}: is not perfect")

perfect(6)
```

global variable

```python
r=100  #global variable
def var():
  print(f"inside {r}")

var()
print(f"outside{r}")
```

local variable

```python

def var():
  i=99
  print(f"inside {i}")

var()
print(f"outside{i}")
```

covert local to global

```python
def var():
  global i
  i=99
  print(f"inside {i}")

var()
print(f"outside {i}")
```

```python
def rev(a):
  c=[]
  i=len(a)-1
  while i>0:
    c.append(a[i])
    i=i-1
  print(c)


rev([1,2,3,4,5,6,7,8,9])
```

```python
#Returning Multiple Values from a Python Function
def cal(num1,num2):
  result1=num1+num2
  result2=num1*num2
  return result1,result2

sum,product=cal(10,20)
print("sum:",sum,"product",product)
```
