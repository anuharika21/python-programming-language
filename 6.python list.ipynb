```python
# remove = pass value it will remove

f = [10,50,90,75]  # save in the Memory
print(f"Before : {f}")

f.remove(90)

print(f"After : {f}") # calling from the Memory
```

```python
# clear = this function will remove all the values
# in the list but varibale will be alive

f = [10,50,90,75]  # save in the Memory
print(f"Before : {f}")

f.clear()

print(f"After : {f}") # calling from the Memory
```

- String Length
- Vowel Count
- Reverse String
- Digit Count
- List Sum
- List Average
- Largest Number
- Smallest Number

```python
text='python123'
numbers=[10,20,30,40,50]

#string len
print("string length:",len(text))

#vowel count
vowels=0
for ch in text.lower():
  if ch in "aeiou":
    vowels +=1
print("vowel count:",vowels)


#reverse
print("reverse:",text[::-1])

#digit count
digit=0
for ch in text:
  if ch.isdigit():
    digit+=1
print("digit count:",digit)


#sum
print("list sum:",sum(numbers))

print("list avg:",sum(numbers)/len(numbers))

print("largest num:",max(numbers))

print("smallest num:",min(numbers))
```

```python
text = "Data Analyst123"

vowels = 0
consonants = 0
digits = 0

for ch in text:
    if ch.lower() in "aeiou":
        vowels += 1
    elif ch.isalpha():
        consonants += 1
    elif ch.isdigit():
        digits += 1

print("Total Characters:", len(text))
print("Total Vowels:", vowels)
print("Total Consonants:", consonants)
print("Total Digits:", digits)
print("Reverse String:", text[::-1])
```
