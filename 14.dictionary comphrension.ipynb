prime

```python
result = {
    i: i*i
    for i in range(2, 21)
    if all(i % j != 0 for j in range(2, i))
}

print(result)
```

```python
keys = ["a", "b", "c"]
values = [10, 20, 30]

result = {k: v for k, v in zip(keys, values)}

print(result)
```
