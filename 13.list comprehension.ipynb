### Dictionary
Create a dictionary

```python
num=[1,2,3,4]
d={i:i*i for i in num}
d
```

swap keys and values

```python
a={'a':1,'b':3}
result={v:k for k,v in a.items()}
result
```

Keep even values only

```python
d = {"a":2,"b":5,"c":8}

result = {k:v for k,v in d.items() if v%2==0}

print(result)
```

square

```python
d = {"a":2,"b":3}

result = {k:v*v for k,v in d.items()}

print(result)
```

upper keys

```python
d = {"a":1,"b":2}

result = {k.upper():v for k,v in d.items()}

print(result)
```

# nested list

```python
lst = [[1,2],[3,4],[5,6]]

result = [j for i in lst for j in i]

print(result)
```

```python
table = [5*i for i in range(1,11)]

print(table)
```

. Cartesian product.

```python
a = [1,2]
b = [3,4]

result = [(i,j) for i in a for j in b]

print(result)
```

Flatten nested strings.

```python
words = [["a","b"],["c","d"]]

result = [j for i in words for j in i]

print(result)
```

Find common elements.

```python
a = [1,2,3,4]
b = [3,4,5]

result = [i for i in a if i in b]

print(result)
```
