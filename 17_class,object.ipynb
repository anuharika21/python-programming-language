add the digits

```python
class ADD:
    def add_digits(self, n):
        try:
            a = 0
            # Convert the number to a string to iterate over each character
            for i in str(n):
                a = a + int(i)  # Convert the character back to an integer to add it
            return a
        except Exception as e:
            return f"An error occurred: {e}"

calculator = ADD()

result = calculator.add_digits(457)
print(result)
```

Palindrome Number

```python
class PALIN:
  def p(self,n):
    try:
      t=n
      s=0

      while n > 0:
        res = n % 10
        s=(s*10)+res
        n=n//10

      if t==s:
        return "palindrome"

      else :
        return "not palindrome"
    except Exception as e:
            er_t, er_m, er_line = sys.exc_info()
            print(f"error in line no: {er_line.tb_lineno} due to: {er_t} reason {er_m}")


t=PALIN()
sol=t.p(111)
print(sol)
```

Amstrong

```python
import sys

class AMSTRONG:
    def amstrong(self, n):
        try:
            t = n
            s = 0

            while n > 0:
                d = n % 10
                s = s + (d * d * d)
                n = n // 10


            if t == s:
                return "amstrong"
            else:
                return "not amstrong"

        except Exception as e:
            er_t, er_m, er_line = sys.exc_info()
            print(f"error in line no: {er_line.tb_lineno} due to: {er_t} reason {er_m}")
am = AMSTRONG()
sol = am.amstrong(153)
print(sol)
```

strong

```python
import sys
class STRONG:
  def strong(self,n):
    try:
      t=n
      s=0
      while n>0:
        d=n%10
        f=1

        for i in range(1,d+1):
          f=f*i
        s=s+f

        n=n//10

      if t==s:
        return 'strong'
      else:
        return "not"
    except Exception as e:
      er_t,er_m,er_line=sys.exc_info()
      print(f"line {er_line.tb_lineno} due to {er_t} reason{er_m}")

s=STRONG()
sol=s.strong(145)
print(sol)
```

perfect

```python
class PERFECT:
  def perfect(self,n):
    try:
      t=0

      for i in range(1,n):
        if n % i==0:
          t+=1

      if  n == t:
        return 'perfect'

      else:
        return 'not perfect'


    except Exception as e:
      er_t,er_m,er_line=sys.exc_info()
      print(f"line {er_line.tb_lineno} due to {er_t} reason{er_m}")
p=PERFECT()
sol=p.perfect(6)
print(sol)
```

gcd

```python
class GCD:
  def gcd(self,a,b):
    try:
      while b:
        a,b=b,a%b
        return abs(a)
    except Exception as e:
      print("error",e)

obj=GCD()
print("GCD",obj.gcd(28,18))
```

```python
import sys
class LEAPYEAR():
  def check(self,year):

    try:
      # A year is a leap year if it is divisible by 400,
            # OR divisible by 4 but NOT by 100
      #return if (year % 400==0) or (year % 100 !=0 and year % 4==0)
          if year % 400 == 0:
            print(f"{year} is a leap year.")
          elif year % 100 == 0:
               print(f"{year} is not a leap year.")
          elif year % 4 == 0:
             print(f"{year} is a leap year.")
          else:
            print(f"{year} is not a leap year.")
    except Exception as e:
      er_t, er_m, er_line = sys.exc_info()
      print(f"error in line no: {er_line.tb_lineno} due to: {er_t} reason {er_m}")


leapyear=LEAPYEAR()
print(leapyear.check(2032))



```

```python
# 8. Password Validator

class PasswordValidator:
    def is_valid(self, password):
        try:
            if len(password) < 8:
                return False

           # has_upper = False
           # has_lower = False
           # has_digit = False

            for ch in password:
                if ch.isupper():
                    has_upper = True
                elif ch.islower():
                    has_lower = True
                elif ch.isdigit():
                    has_digit = True

            return has_upper and has_lower and has_digit

        except Exception as e:
            print("Error:", e)
            return False


obj = PasswordValidator()

pwd = input("Enter a password: ")
if obj.is_valid(pwd):
    print("Valid password")
else:
    print("Invalid password")
```
