```python
#Write a program using OR operator to check login conditions.
# Simulated database of a registered user
correct_username = "coder123"
correct_email = "coder@example.com"
correct_password = "SecurePassword7"

# Simulated user input (Imagine these came from a login form)
user_identifier = input("Enter your username or email: ")
user_password = input("Enter your password: ")

# Check the conditions using the OR operator
# The user passes the first check if they provide the correct username OR the correct email
if (user_identifier == correct_username or user_identifier == correct_email) and user_password == correct_password:
    print("Login successful! Welcome back.")
else:
    print("Login failed. Invalid credentials.")
```

```python
# The authorized master PINs
admin_pin_1 = "1234"
admin_pin_2 = "9999"

# Take input from the user
entered_pin = input("Enter Admin PIN to log in: ")

# Check the condition using ONLY the 'or' operator
if entered_pin == admin_pin_1 or entered_pin == admin_pin_2:
    print("Login successful! Access granted.")
else:
    print("Login failed. Invalid PIN.")
```
