```python
marks = {
    "Math": 95,
    "Science": 82,
    "English": 91
}

highest_subject = ""
highest_mark = 0

for subject, mark in marks.items():
    if mark > highest_mark:
        highest_mark = mark
        highest_subject = subject

print("Highest Subject:", highest_subject)
print("Highest Mark:", highest_mark)
```
