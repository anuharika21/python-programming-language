
- `Python set concept is going to represent with -> {} -> anything we write inside the {} it becomes set `

- `Python Set concept is completely an Unordered Pair `

- `In Python set, there will be no index concept. If there is no index, slicing and skipping concept also are not at all possible.`

- `Python Set concept will not allow duplicates `

- `Inside the Python Set concept we can pass anydata type except Mutable

```python
f={10,30.3,'data',(1,2,3)}
print(f)
```

```python
f = {10 , 15.55 , 'data' , (1,2,3)} # save in the Memory

print(f) # calling
```

```python
f = {10 , 15.55 , 'data' , [1,2,3]} # save in the Memory

print(f) # calling
```


**Python Set Built_in_Functions**

**Increasing the Memory Size**

- `add`
- `update `

**Decreasing the Memory size**
- `pop `
- `remove `
- `discard `
- `clear `

**Some Common Operations**
- `copy `
- `union `
- `intersection `


```python
# add = this function will take only 1 input from user and will add to the existing set
# randomlly

a = {10,20,30,40,50} # save in the Memory
print(f"Before : {a}")

a.add(500)

print(f"After : {a}")
```

```python
# update = this function will take mutiple inputs from user and will add to the existing set
# randomlly

a = {10,20,30,40,50} # save in the Memory
print(f"Before : {a}")

a.update([500,600,900])

print(f"After : {a}")
```

```python
# pop in set concept will remove one value randomlly
f = {10,20,50,90,100}  # save in the Memory
print(f"Before : {f}")

f.pop()

print(f"After : {f}")
```

```python
# discard = this function will take the value from the user and if the value is present
# it will remove and if the value is not present it will be silent


f = {10,20,50,90,100}  # save in the Memory
print(f"Before : {f}")

f.discard(50)

print(f"After : {f}")
```

```python
# discard = this function will take the value from the user and if the value is present
# it will remove and if the value is not present it will be silent


f = {10,20,50,90,100}  # save in the Memory
print(f"Before : {f}")

f.discard(50)

print(f"After : {f}")
```

```python
# remo = this function will take the value from the user and if the value is present
# it will remove and if the value is not present it will through error


f = {10,20,50,90,100}  # save in the Memory
print(f"Before : {f}")

f.remove(500)

print(f"After : {f}")
```

```python
# clear

t = {10,20,30}

t.clear()

print(t)
```

```python
c = []
print(f"Before : {c}")


c.append(10)
c.append(20)

print(f"After : {c}")
```

```python
t = {10,20,30}
p = t # shallow

print(t)
print(p)

print(id(t))
print(id(p))

t.add(100)

print(t)
print(p)
```

```python
t = {10,20,30}
p = t.copy() # deep

print(t)
print(p)

print(id(t))
print(id(p))

t.add(100)

print(t)
print(p)
```
